# Wrangling Scraped Data 
Since the data in *guitars.csv* was scraped from eBay, it has a lot of incorrectly entered values, missing values, bad spellings etc. In this notebook, I've documented the highlights from turning this scraped csv with tens of thousands of missing or incorrect values into something suitable for fitting the linear model in this project. Some of this code will be repeated in the main notebook.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = '../data/guitars.csv'
guitars = pd.read_csv(data)
guitars.shape
```

## Some Initial Issues
The original intent was to collect all of the ~18k samples in the eBay sold and completed electric guitar listing search. Due to some time and technical constraints, the scraper was only able to collect 12,455 with 37 columns, many of which had more than 10k missing values. Among some of the most problematic details of the dataset:
- Some extreme outliers (guitars prices anywhere 
  from 10k to 17k, usually signed by artist)
- Entries submitted to an incorrect field
- Variations on the same entry ("Right-Handed", 
  "Right Hand", "Right-Hand Player" for example)
- Multiple columns for the same feature, but some 
  corresponding columns contained the values that 
  could fill another's nulls
- Long, unformatted strings 
- A sizeable quantity of missing values

### Some Helpful Functions
In order to combat the problematic data set, I used a few functions to help produce some reccomendations:`

```python
# Some useful functions
def findHoles(dataframe, a, b):
    '''Returns instances where the left dataframe 
       in question has nulls that the right dataframe 
       could fill'''
    condition = dataframe[a].isnull() & ~dataframe[b].isnull()
    return dataframe.loc[condition, [a, b]]

def checkTwin(dataframe, a, b):
    '''Returns part of the dataframe
       where values are identical'''
    return dataframe.loc[dataframe[a] == dataframe[b]][[a, b]]

def mergeCheck(dataframe):
    '''Recommend columns to merge'''
    cache = []
    for x in range(len(dataframe.columns)):
        for y in range(len(dataframe.columns)):
            if y == x:
                continue
            elif (x, y) in cache or (y, x) in cache:
                continue
            else:
                feature1 = dataframe.columns[x]
                feature2 = dataframe.columns[y]
                result = len(dataframe.loc[dataframe[feature1] == dataframe[feature2]])
                if result == 0:
                    cache.append((x, y))
                    continue
                elif result > 50:
                    f1_null = dataframe[feature1].isnull().sum()
                    f2_null = dataframe[feature2].isnull().sum()
                    
                    if f1_null > f2_null:
                        most_nulls = (f1_null, feature1)
                    else:
                        most_nulls = (f2_null, feature2)
                    print(f"Consider merging:\n\t{dataframe.columns[x]} "\
                          f"and {dataframe.columns[y]}\n\t{result} matching rows\n\t"\
                          f"Null values in {most_nulls[1]}: {most_nulls[0]}")
                    cache.append((x, y))
                else:
                    continue
```

## Taking Action
The following function is one that I plan on repeating in the main notebook, plus or minus a few modifications. I attempted to correct some of the most glaring issues in the dataset
- As far as predicting price, UPC is only of interest insofar as it relates to influencing the buyer's decision to commit to the asking or bidding price. This should preferably be a binary classification.
- Seller confusion caused many MPN (manufacturer product number or part number) to be entered incorrectly. This column is another canditate to be transformed to a two class category.

***Note:*** *Adjusting the string configuration to the extent seen here probably will not improve the final model's score by any significant measure, or help with reproducibilty on new data, rather, it was done in the interest of assuaging my personal compulsions 🙃*

```python
# Remove crazy outliers above $10,000
guitars = guitars.drop(index=guitars.loc[guitars['Price'] > 9999.99].index)
```

```python
def wrangle(X):
    '''Bulk wrangle for split data'''
    
    # Avoid setting with copy
    X = X.copy()
    
    # Set feature thresholds
    model_limit = 50
    brand_limit = 100
    color_limit = 100
    material_limit = 50
    btype_limit = 30
    size_limit = 10
    
    # Clean conditions column text
    X['Condition'] = X['Condition'].str.replace('\:\D+', '')
    
    # Make null values out of the incorrectly entered categories
    conditions = ['Used', 'New', 'Open box', 'Seller refurbished',
                  'For parts or not working', 'Manufacturer refurbished']
    
    X.loc[~X['Condition'].isin(conditions), 'Condition'] = np.nan
    
    # Combine columns that are obviously the same and where values are null
    to_merge = {'Right-/ Left-Handed': ['Dexterity', 'Hand'],
                'Country/Region of Manufacture': ['Country of Manufacture'], 
                'Body Color': ['Color'], 'Body Type': ['Type']}
    for main_feat, dup_feat in to_merge.items():
        for f in dup_feat:
            condition = X[main_feat].isnull() & ~X[f].isnull()
            X.loc[condition, main_feat] = X[f]
            # Drop column after merge
        X = X.drop(columns=dup_feat)
        
    # Reduce categories in the right-/ left hand column
    X['Right-/ Left-Handed'] = X['Right-/ Left-Handed'].str.replace(
    '(right or left|left or right|either|lap|right-handed,).*', 
    'both')
    X['Right-/ Left-Handed'] = X['Right-/ Left-Handed'].str.replace(
    'r.*', 'right-handed')
    X['Right-/ Left-Handed'] = X['Right-/ Left-Handed'].str.replace(
    'l.*', 'left-handed')
    
    # Remove names that aren't real models from top 100
    X['Model'] = X['Model'].str.upper()
    not_a_model = ['DRN979CTV LH BK 2ND', '39" ELECTRIC GUITAR', '2019',
                   'DOES NOT APPLY', 'UNKNOWN', 'GTR235 LH BK 2ND']  
    top_models = X['Model'].value_counts()[:model_limit].index
    
    X.loc[X['Model'].isin(not_a_model), 'Model'] == 'OTHER'
    X.loc[~X['Model'].isin(top_models), 'Model'] == 'OTHER'
    
    # Clean and shrink the brand column
    X['Brand'] = X['Brand'].str.upper()
    X['Brand'] = X['Brand'].str.replace('UNBRANDED.*', 'UNBRANDED')
    top_brands = X['Brand'].value_counts()[:brand_limit].index
    X.loc[~X['Brand'].isin(top_brands), 'BRAND'] == 'OTHER'
    
    # Clean and shrink cardinality of the string configuration column
    standard_configs = ['6 String', '7 String', '8 String', '9 String', '12 String']
    
    X['String Configuration'] = X['String Configuration'].str.upper()
    X['String Configuration'] = X['String Configuration'].str.replace(
        '6\W*STRING(?!78912).*|^6$|6\s*\w*\s*STRINGS(?!78912)', '6 STRING')
    X['String Configuration'] = X['String Configuration'].str.replace('^7$', '7 STRING')
    X['String Configuration'] = X['String Configuration'].str.replace('12\W*STRING', '12 STRING')
    
    X.loc[~X['String Configuration'].isin(standard_configs), 'String Configuration'] == 'OTHER'
    
    # Turn UPC and MPN into included and not included categories
    X['UPC'] = X['UPC'].str.upper()
    X.loc[X['UPC'] != 'DOES NOT APPLY', 'UPC'] = 'PROVIDED'
    X.loc[X['UPC'] == 'DOES NOT APPLY', 'UPC'] = 'NOT AVAILABLE'
    
    # If MPN entry contains a digit, it can be considered a valid entry
    mpn_condition = X['MPN'].str.contains('\d')
    X.loc[mpn_condition == True, 'MPN'] = 'PROVIDED'
    X.loc[mpn_condition == False, 'MPN'] = 'NOT AVAILABLE'
    
    # Reduce size, color and material cardinality
    features_limits = {'Body Color': color_limit,
                       'Body Material': material_limit,
                       'Size': size_limit}
    
    for feat, lim in features_limits.items(): # Apply the same wrangling to 3 features
        X[feat] = X[feat].str.upper()
        top = X[feat].value_counts()[:lim].index
        X.loc[~X[feat].isin(top), feat] = 'OTHER'
    
    # Reduce body type cardinality
    X['Body Type'] = X['Body Type'].str.upper()
    top_btypes = X['Body Type'].value_counts().index[:btype_limit]
    X['Body Type'].str.replace('STRAT.*', 'STRATOCASTER')
    X.loc[~X['Body Type'].isin(top_btypes), 'Body Type'] = 'OTHER'
    
    # Make model year a numeric category
    X.loc[X['Model Year'].str.contains('^\d{4}$') == False, 'Model Year'] = np.nan
    X['Model Year'] = pd.to_numeric(X['Model Year'])    
    
    return X
```

```python
guitars = wrangle(guitars)
```
